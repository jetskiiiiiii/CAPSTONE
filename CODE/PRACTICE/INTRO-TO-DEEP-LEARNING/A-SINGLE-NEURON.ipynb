# The Linear Unit

For an input *x*, its connection to the neuron has a weight *w*.

When a value flows through a connection, what reaches the neuron is *w\*x*.

<br/>

A neuron also has a special weight connected to it called bias *b*.

The bias has no input (just 1).

<br/>

For an ouput *y*, a neuron sums up all the values it receives through its connections (e.g. *y=mx+b*).

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
import matplotlib.pyplot as plt
```

```python
# model is not trained, so weight is random and bias is 0
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[1]) # 'units' define the number of outputs, 'input_shape' defines the number of features (columns in data)
])

x = tf.linspace(-1, 1, 100) # creating 100 inputs
y = model.predict(x) # creates outputs based on the random weight

plt.figure(dpi=100)
plt.plot(x, y, 'k')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Input: x")
plt.ylabel("Target y")
w, b = model.weights # you could also use model.get_weights() here
plt.title("Weight: {:0.2f}\nBias: {:0.2f}".format(w[0][0], b[0]))
plt.show()
```

```python
import pandas as pd

# Red Wine dataset has data on 1600 wines
red_wine_data = pd.read_csv('./red-wine.csv')
red_wine_data.head()
```

```python
# the output is the 'qualtiy'
# the input/features determine the input_shape

input_shape = [red_wine_data.shape[1] - 1] # enclosed in brackets; # of columns - 1 to account for the ouput column

# setting the input shape also tells Keras the dimensions of each example of the training data (in this case [1, 12])
```

```python
# implementing a model
red_wine_model = keras.Sequential([
    layers.Dense(units=1, input_shape=input_shape)
])
```

```python
# get the initial weights and bias
weights, bias = red_wine_model.weights
print(f"Weights\n{weights}\nBias\n{bias}")

# one weight for one input, plus a bias
# weights are initially set randomly (and bias set to 0), until trained - at which point the model will try to find better values for its weights
```
